## Q5  Propose two hypothesis tests exploring information related to user loyalty score. For each hypothesis test, you must describe the motivation (why this hypothesis is interesting and important to test), null hypothesis, and select a proper statistical test to report the test result.

- *Hypothesis Test 1*:


    - Motivation: The first hypothesis test could explore whether there is a significant difference in user loyalty scores  between customers who have made purchases from merchants in different merchant categories. This hypothesis is interesting because it could help merchants to understand how their merchant category affects customer loyalty, and could inform marketing strategies to improve customer loyalty.
    

    - Null Hypothesis: There is no significant difference in user loyalty scores between customers who have made purchases
      from merchants in different merchant categories.

```python
# Separate the data into two groups based on the values in the 'Merchant Group ID' column

high_loality = total_data[total_data['merchant_group_id'] > 19492]['score']
low_loality = total_data[total_data['merchant_group_id'] <= 19492]['score']

# Perform the Mann-Whitney U test
statistic, p_value = mannwhitneyu(high_loality, low_loality)
print('Mann-Whitney U statistic:', statistic)
print('p-value:', p_value)

# Check if the p-value is less than the significance level of 0.05
if p_value < 0.05:
    print('The difference in loyalty scores between the two groups is statistically significant.')
else:
    print('The difference in loyalty scores between the two groups is not statistically significant.')
```

- because p_valueis less than 0.05 and the result is "The difference in loyalty scores between the two groups is statistically significant," it appears that the null hypothesis has been rejected, and there is a significant difference in loyalty scores between the two groups so i keep this column to learn model with it.
